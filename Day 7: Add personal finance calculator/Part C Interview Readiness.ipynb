```python
print(type(True))  # This returns <class 'bool'> because True is a Boolean data type in Python.

print(isinstance(True, int))  
# This returns True because in Python, bool is a subclass of int.
# Internally, True behaves like 1 and False behaves like 0.

print(True + True + False)  
# This returns 2 because True = 1 and False = 0.
# So 1 + 1 + 0 = 2.

print(int(3.99))  
# This returns 3 because int() truncates the decimal part.
# It does NOT round the number.

print(bool("False"))  
# This returns True because any non-empty string in Python evaluates to True.
# Even the string "False" is still non-empty, so it is truthy.

print(bool(""))  
# This returns False because an empty string is considered False in Python.
# Empty objects are treated as False.

print(0.1 + 0.2 == 0.3)  
# This returns False due to floating-point precision issues.
# 0.1 + 0.2 internally becomes 0.30000000000000004,
# which is not exactly equal to 0.3.

print("5" + "3")  
# This returns "53" because both values are strings.
# The + operator performs string concatenation.

print(5 + 3)  
# This returns 8 because both values are integers.
# The + operator performs arithmetic addition.
```

```python
def analyze_value(value):
    # Get the type name of the value
    value_type = type(value).__name__
    
    # Determine truthiness (True or False)
    truthy = bool(value)
    
    # Try to get length (only works for iterable types)
    try:
        length = len(value)
    except TypeError:
        length = "N/A"
    
    # Return formatted string
    return (
        f"Value: {value} | "
        f"Type: {value_type} | "
        f"Truthy: {truthy} | "
        f"Length: {length}"
    )
```

```python
name = input("Name: ")
age = int(input("Age: "))

if age >= 18:
    status = "Adult"
else:
    status = "Minor"

print(f"{name} is {age} years old and is a {status}")
print(f"In 5 years: {age + 5}")

score = 85.5
print(f"Score: {score:.0f}")
```
