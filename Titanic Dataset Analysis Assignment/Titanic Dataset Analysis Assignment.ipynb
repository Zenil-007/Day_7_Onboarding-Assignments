🧠 CELL 2 – Load CSV Manually (No Pandas)

Loading CSV manually (No Pandas)

1. Imported CSV module to process csv files. 
2. Taking the data in a list as a reader
3. Creating a new variable named data and storing reader values as a list. 


```python
with open(file_path, "r") as file:
    reader = csv.reader(file)
    header = next(reader)
    data = list(reader)

print("Total Rows:", len(data))
print("Columns:", header)
```

🧠 CELL 3 – Convert Age Column to Python List

```python
age_index = header.index("Age")

age_list = []
for row in data:
    if row[age_index] != "":
        age_list.append(float(row[age_index]))

print("Total Valid Ages:", len(age_list))
```

🧠 CELL 4 – Mean, Median, Std Dev (NumPy)



```python
age_array = np.array(age_list)

mean_age = np.mean(age_array)
median_age = np.median(age_array)
std_age = np.std(age_array)

print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Std Dev Age:", std_age)
```

🧠 CELL 5 – Fare Array + Top/Bottom 10%

```python
fare_index = header.index("Fare")

fare_list = []
for row in data:
    fare_list.append(float(row[fare_index]))

fare_array = np.array(fare_list)

top_10_threshold = np.percentile(fare_array, 90)
bottom_10_threshold = np.percentile(fare_array, 10)

top_10 = fare_array[fare_array >= top_10_threshold]
bottom_10 = fare_array[fare_array <= bottom_10_threshold]

print("Top 10% Fare Count:", len(top_10))
print("Bottom 10% Fare Count:", len(bottom_10))
```

🧠 CELL 6 – Survival by Age Groups (NumPy Filtering)

```python
survived_index = header.index("Survived")

age_group = []
survived = []

for row in data:
    if row[age_index] != "":
        age_group.append(float(row[age_index]))
        survived.append(int(row[survived_index]))

age_group = np.array(age_group)
survived = np.array(survived)

child_survival = survived[age_group < 15].mean()
adult_survival = survived[(age_group >= 15) & (age_group <= 60)].mean()
senior_survival = survived[age_group > 60].mean()

print("Child Survival:", child_survival)
print("Adult Survival:", adult_survival)
print("Senior Survival:", senior_survival)
```

🧠 CELL 7 – Load With Pandas

```python
df = pd.read_csv(file_path)
df.head()
```

🧠 CELL 8 – Handle Missing Values

```python
df["Age"] = df.groupby("Pclass")["Age"].transform(lambda x: x.fillna(x.median()))
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

🧠 CELL 9 – Feature Engineering

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = np.where(df["FamilySize"] == 1, 1, 0)
df["FarePerPerson"] = df["Fare"] / df["FamilySize"]
```

🧠 CELL 10 – Categorical Binning

```python
df["AgeGroup"] = pd.cut(df["Age"],
                        bins=[0,15,60,100],
                        labels=["Child","Adult","Senior"])

df["FareGroup"] = pd.qcut(df["Fare"],
                          q=3,
                          labels=["Low","Medium","High"])
```

🧠 CELL 14 – Encode Variables

```python
df["GenderNum"] = np.where(df["Sex"] == "female", 1, 0)
df["PclassNorm"] = 1 - (df["Pclass"] - 1)/2
```

🧠 CELL 15 – Normalize Numeric Columns

```python
def normalize(x):
    return (x - x.min()) / (x.max() - x.min())

df["AgeNorm"] = normalize(df["Age"])
df["FareNorm"] = normalize(df["Fare"])
df["FamilyNorm"] = normalize(df["FamilySize"])
```

🧠 CELL 16 – Weighted Survival Score

```python
w1 = 0.4  # Gender
w2 = 0.2  # Pclass
w3 = 0.1  # Age
w4 = 0.2  # Fare
w5 = 0.1  # Family

df["SurvivalScore"] = (
    w1*df["GenderNum"] +
    w2*df["PclassNorm"] +
    w3*(1-df["AgeNorm"]) +
    w4*df["FareNorm"] +
    w5*(1-df["FamilyNorm"])
)

df["Prediction"] = np.where(df["SurvivalScore"] >= 0.5, 1, 0)
```

🧠 CELL 17 – Evaluation Metrics (Manual NumPy)

```python
y_true = df["Survived"].values
y_pred = df["Prediction"].values

TP = np.sum((y_true == 1) & (y_pred == 1))
TN = np.sum((y_true == 0) & (y_pred == 0))
FP = np.sum((y_true == 0) & (y_pred == 1))
FN = np.sum((y_true == 1) & (y_pred == 0))

accuracy = (TP + TN) / len(y_true)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)

print("\nConfusion Matrix")
print(np.array([[TP, FP],
                [FN, TN]]))
```

📊 Insight (Markdown)

1. Model significantly outperforms random guessing (50% baseline).
2. Manual feature-weight logic captures structured survival bias.

1. Priority Rescue Profile:
   Women, children, first-class passengers.

2. Ethical Concerns:

🧠 CELL 19 – Predict Function

```python
def predict_survival(passenger):
    gender = 1 if passenger["Sex"] == "female" else 0
    pclass = 1 - (passenger["Pclass"] - 1)/2
    
    score = (
        0.4*gender +
        0.2*pclass +
        0.1*(1-passenger["AgeNorm"]) +
        0.2*passenger["FareNorm"] +
        0.1*(1-passenger["FamilyNorm"])
    )
    
    return 1 if score >= 0.5 else 0
```
